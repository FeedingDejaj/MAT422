## Graph Definitions

### Key Terms:
- **Vertex (Node)**: Fundamental unit of a graph representing an object.
- **Edge**: Connection between two vertices, which can be directed or undirected.
- **Weighted Graph**: Graph where edges have associated weights.
- **Adjacency Matrix**: A 2D array representing the connections between nodes.
- **Degree**: Number of edges connected to a vertex (in-degree and out-degree for directed graphs).

Graphs can be categorized into:
- **Undirected Graphs**: Edges have no direction.
- **Directed Graphs (Digraphs)**: Edges have a direction.
- **Weighted Graphs**: Edges have weights (e.g., distances, costs).


```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()


G.add_nodes_from([1, 2, 3, 4])


G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Draw the graph
plt.figure(figsize=(8, 5))
nx.draw(G, with_labels=True, node_color='skyblue', edge_color='black', node_size=1500, font_size=15)
plt.title('Undirected Graph')
plt.show()
```

## Graph Modeling

Graph modeling is the process of representing real-world problems using graphs. Examples include:
- **Social Networks**: Nodes represent people, and edges represent friendships or connections.
- **Transportation Networks**: Nodes represent locations, and edges represent routes or roads.
- **Biological Networks**: Nodes represent proteins or genes, and edges represent interactions.
- **Web Graph**: Nodes represent webpages, and edges represent hyperlinks.

```python
import networkx as nx
import matplotlib.pyplot as plt


social_network = nx.DiGraph()


social_network.add_nodes_from(['Alice', 'Bob', 'Charlie', 'Diana'])


social_network.add_edges_from([
    ('Alice', 'Bob'),
    ('Bob', 'Charlie'),
    ('Charlie', 'Alice'),
    ('Diana', 'Alice')
])


plt.figure(figsize=(8, 5))
nx.draw(social_network, with_labels=True, node_color='lightgreen', edge_color='gray',
        node_size=2000, font_size=15, arrowsize=20)
plt.title('Social Network Graph')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt


weighted_graph = nx.Graph()


weighted_graph.add_nodes_from([1, 2, 3, 4])


weighted_graph.add_weighted_edges_from([
    (1, 2, 4.5),
    (2, 3, 3.0),
    (3, 4, 2.1),
    (4, 1, 1.8)
])


pos = nx.spring_layout(weighted_graph)
plt.figure(figsize=(8, 5))
nx.draw(weighted_graph, pos, with_labels=True, node_color='orange', edge_color='black',
        node_size=1500, font_size=15)
edge_labels = nx.get_edge_attributes(weighted_graph, 'weight')
nx.draw_networkx_edge_labels(weighted_graph, pos, edge_labels=edge_labels, font_size=12)
plt.title('Weighted Graph')
plt.show()
```
