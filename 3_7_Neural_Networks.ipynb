## 3.7.1 Mathematical Formulation

A **neural network** is a computational model inspired by the human brain. It consists of layers of neurons connected by weighted edges. The main components include:

- **Input layer**: Takes input features.
- **Hidden layers**: Perform computations using weights, biases, and activation functions.
- **Output layer**: Provides the final prediction or classification.

Mathematically, the output of a neuron is:
\[
z = W \cdot X + b
\]
where:
- \(W\): Weight matrix
- \(X\): Input vector
- \(b\): Bias term

The activation function \(g(z)\) transforms \(z\) into the neuron's output.

## 3.7.2 Activation Functions

Activation functions introduce non-linearity into the network, allowing it to learn complex patterns. Common activation functions include:

```python
import numpy as np
import matplotlib.pyplot as plt


def step(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-10, 10, 100)
y = step(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Step Function')
plt.title('Step Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
def positive_linear(x):
    return np.maximum(0, x)

y = positive_linear(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Positive Linear Function (ReLU)')
plt.title('Positive Linear Function (ReLU)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y = sigmoid(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sigmoid Function')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

x = np.array([2.0, 1.0, 0.1])
y = softmax(x)

plt.figure(figsize=(8, 5))
plt.bar(range(len(y)), y, label='Softmax Output')
plt.title('Softmax Function')
plt.xlabel('Classes')
plt.ylabel('Probability')
plt.grid()
plt.legend()
plt.show()
```

## 3.7.3 Cost Function

The **cost function** measures the error between predicted and actual outputs. For neural networks, common cost functions include:

- **Mean Squared Error (MSE)**: Used for regression tasks.
- **Cross-Entropy Loss**: Used for classification tasks.

Cross-Entropy Loss is defined as:
\[
L = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1-y_i) \log(1-\hat{y}_i) \right]
\]

## 3.7.4 Backpropagation

Backpropagation is an algorithm used to compute gradients of the cost function with respect to network parameters (weights and biases). It propagates errors backward through the network, updating parameters to minimize the cost function.

## 3.7.5 Backpropagation Algorithm

The backpropagation algorithm involves:

1. **Forward Pass**: Compute the network's output and the cost function.
2. **Backward Pass**: Calculate the gradient of the cost function using chain rule.
3. **Parameter Update**: Update weights and biases using gradient descent:
\[
w = w - \eta \frac{\partial L}{\partial w}, \quad b = b - \eta \frac{\partial L}{\partial b}
\]
where \(\eta\) is the learning rate.
