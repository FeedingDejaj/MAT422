## 3.2.1 Limits and Continuity

**Limits** and **continuity** are foundational concepts in calculus that help describe the behavior of functions as inputs approach certain values.

### Topics Covered:
- **Limits of a Function**: The value a function approaches as its input gets closer to a specific point.
- **Continuous Functions**: Functions that are smooth and unbroken over an interval.
- **Extremum**: The maximum or minimum values a function can attain within a given range.

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x):
    if x < 2:
        return x**2
    elif x == 2:
        return None
    else:
        return 4 + (x - 2)


x_vals = np.linspace(0, 4, 100)
y_vals = [f(x) if f(x) is not None else np.nan for x in x_vals]


plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.axvline(2, color='red', linestyle='--', label='x=2 (Discontinuity)')
plt.title('Example of a Function with a Limit')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

### Continuous Functions

A function is **continuous** if it is defined and smooth at every point within an interval. This means there are no breaks, jumps, or undefined points in the function's graph. Mathematically, a function \(f(x)\) is continuous at \(x = a\) if:

1. f(a) is defined.
2. The limit lim(x to a) f(x) exists.
3. lim(x to a) f(x) = f(a).

### Extremum

An **extremum** is a point at which a function reaches its maximum or minimum value. These can occur as:

- **Local extremum**: Maximum or minimum values within a specific interval.
- **Global extremum**: The overall maximum or minimum value over the entire domain of the function.

Extremum points are often found by setting the derivative of the function to zero and solving for the critical points.

## 3.2.2 Derivatives

The **derivative** is a measure of how a function changes as its input changes. It represents the slope of the function at any given point.

### Topics Covered:
- **Single-Variable Case**: Understanding the basic derivative.
- **Partial Derivatives**: Derivatives with respect to one variable while holding others constant.
- **Jacobian Matrix**: A matrix of all first-order partial derivatives for vector-valued functions.

```python
from sympy import symbols, diff


x = symbols('x')
f = x**2 + 3*x + 2


f_prime = diff(f, x)
f_prime
```

```python
from sympy import symbols, Matrix


x, y = symbols('x y')
f1 = x**2 + y
f2 = x + y**2


F = Matrix([f1, f2])


J = F.jacobian([x, y])
J
```

## 3.2.3 Taylor’s Theorem

**Taylor’s Theorem** approximates a function near a given point using polynomials. The approximation improves as more terms are added.

For a function \(f(x)\) centered at \(a\), the Taylor series is:

\[f(x) \approx f(a) + f'(a)(x-a) + \frac{f''(a)}{2!}(x-a)^2 + \ldots\]

```python
from sympy import symbols, diff, factorial, exp


x = symbols('x')
f = exp(x)


a = 0
n_terms = 5
taylor_series = sum(diff(f, x, n).subs(x, a) / factorial(n) * (x - a)**n for n in range(n_terms))
taylor_series
```
